**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.**

1)Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used in the scientific and data analysis communities to generate various types of plots and charts, allowing users to visualize data in an easy-to-understand format.

2)Matplotlib is used for several reasons:
    
1. **Versatility**: It can produce a wide range of plots and charts, from simple line graphs to complex 3D plots.

2. **Customization**: It provides a high level of control over plot appearance, including fonts, colors, and line styles.

3. **Integration**: It works well with other libraries such as NumPy and Pandas, making it a powerful tool in the data science and scientific computing ecosystem.

4. **Publication Quality**: It can produce figures that are suitable for publication in scientific journals and presentations.

5. **Interactivity**: Through tools like Jupyter Notebooks, Matplotlib allows for interactive data exploration and visualization.

3)The Pyplot module is a collection of functions that make Matplotlib work like MATLAB, allowing users to create plots with ease. Here are five types of plots that can be created using Pyplot:



1. **Line Plot**
  
 A line plot is used to display information as a series of data points called 'markers' connected by straight line segment

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.show()
```

2.**Scatter Plot**

A scatter plot uses dots to represent the values obtained for two different variables - one plotted along the x-axis and the other plotted along the y-axis.


```python
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30])
plt.show()
```

3.**Bar Plot**

A bar plot presents categorical data with rectangular bars, with lengths proportional to the values that they represent.

```python
plt.bar(['A', 'B', 'C', 'D'], [10, 20, 25, 30])
plt.show()
```

4.**Histogram**

A histogram is used to represent the distribution of numerical data by dividing it into bins and counting the number of observations in each bin.

```python
data = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5]
plt.hist(data, bins=5)
plt.show()
```

5.**Pie Chart**

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportions.

```python
labels = 'A', 'B', 'C', 'D'
sizes = [10, 20, 25, 30]
plt.pie(sizes, labels=labels)
plt.show()
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:**

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Why is the `subplot()` Function Used?

The `subplot()` function in Matplotlib is used to create multiple plots in a single figure. This allows for a more organized and efficient way to present multiple graphs, especially when comparing them side by side. By using `subplot()`, you can define a grid of subplots and specify the position of each individual plot within this grid.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y1, 'r')  # 'r' is the color red
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x, y2, 'g')  # 'g' is the color green
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x, y3, 'b')  # 'b' is the color blue
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x, y4, 'm')  # 'm' is the color magenta
axs[1, 1].set_title('Line 4')

plt.tight_layout()
plt.show()
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

What is a Bar Plot?

A bar plot (or bar chart) is a graphical display of data using bars of different heights. It is used to compare different categories of data. Each bar represents a category, and the length or height of the bar is proportional to the values they represent.

 Why is it Used?

Bar plots are used because they:
1. **Compare Data Across Categories**: They make it easy to compare quantities among different groups.
2. **Show Trends Over Time**: When used with time series data, bar plots can show trends or changes over time.
3. **Highlight Differences**: They make differences between categories visually obvious.
4. **Simple to Understand**: Bar plots are straightforward and can be understood by a wide audience.


```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profit ")
```

```python
plt.barh(company,profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profit - Horizontal Bar Plot")
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

 What is a Box Plot?

A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It may also display outliers.

 Why is it Used?

Box plots are used because they:
1. **Summarize Data**: Provide a concise summary of the distribution and variability of the data.
2. **Compare Distributions**: Allow easy comparison of distributions between different groups.
3. **Identify Outliers**: Highlight potential outliers in the data.
4. **Show Skewness**: Indicate the skewness of the data distribution.



```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box1', 'Box2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box1 and Box2')
plt.show()
```
